# Beginner Level Task 4 - Image to Pencil Sketch with Python

```python
import cv2
from matplotlib import pyplot as plt
```

```python
#read the image
image = cv2.imread('image1.jpg')
```

```python
#to conver colored image into grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#to invert the grayscale image
inverted_image = 255 - gray_image
```

```python
#blure the inverted image using GaussianBlur function
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
#now inverting that blured image
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
#this is an Original Image
plt.figure(figsize=(15,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

```python
#This is pencil sketch Image
plt.figure(figsize=(15,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.show()
```
